```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('./People.jpg')
plt.imshow(img[:, :, ::-1])
```

```python
img_crop=img[100: 220, 140: 210]
plt.imshow(img_crop[:, :, ::-1])
```

```python
lower_red=np.array([0, 0, 100])
upper_red=np.array([45, 45, 255])

img_binary=cv2.inRange(img_crop, lower_red, upper_red)

plt.imshow(img_binary, 'gray')
```

```python
mask=np.zeros((274, 392))
mask[100: 220, 140: 210]=img_binary
plt.imshow(mask, 'gray')
```

```python
img[:, :, -1]=mask
plt.imshow(img[:, :, ::-1])
```
