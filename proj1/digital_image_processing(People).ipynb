```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('./People.jpg')
plt.imshow(img[:, :, ::-1])
```

```python
lower_red=np.array([0, 0, 100])
upper_red=np.array([45, 45, 255])

img_binary=cv2.inRange(img, lower_red, upper_red)

plt.imshow(img_binary, 'gray')
```

```python
img[:, :, 2]=img_binary
plt.imshow(img[:, :, ::-1])
```
