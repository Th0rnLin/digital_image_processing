```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('./Coin.jpg')
plt.imshow(img)
```

```python
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, 'gray')
```

```python
import numpy as np

plt.hist(img_gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
img_equalize=cv2.equalizeHist(img_gray) 
plt.imshow(img_equalize, 'gray')
```

```python
import numpy as np

plt.hist(img_equalize.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
_, img_threshold=cv2.threshold(img_equalize, 230, 255, cv2.THRESH_BINARY)
plt.imshow(img_threshold, 'gray')
```

```python
img_blur=cv2.GaussianBlur(img_threshold, (23, 23), 0)
plt.imshow(img_blur, 'gray')
```

```python
img_canny=cv2.Canny(img_blur, 160, 160)
plt.imshow(img_canny, 'gray')
```

```python
coin=img_gray.copy()
cnt, _=cv2.findContours(img_canny, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(coin, cnt, -1, 0, 2)
plt.imshow(coin, 'gray')
```
