```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('./Coin.jpg')
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_equalize=cv2.equalizeHist(img_gray) 
_, img_threshold=cv2.threshold(img_equalize, 230, 255, cv2.THRESH_BINARY)
img_blur=cv2.bilateralFilter(img_threshold, 10, 100, 100)
img_canny=cv2.Canny(img_blur, 160, 160)

coin=img_gray.copy()
cnt, _=cv2.findContours(img_canny, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(coin, cnt, -1, 0, 2)
plt.imshow(coin, 'gray')
```
